# Démonstration d'un _clustering_ k-means avec `scikit-learn`

## Première partie : les données

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=60, centers=3, n_features=2, random_state=0)
```

**TODO: ajouter ici une prévisualisation des données brutes.**

## Deuxième partie : clustering

```python
from sklearn.cluster import KMeans

k_means = KMeans(init="k-means++", n_clusters=3, n_init=10)
k_means.fit(X)
```

**TODO: ci-dessous utiliser des couleurs plus jolies.**

```python
import matplotlib.pyplot as plt

colors = ["red", "green", "blue"]

plt.scatter(X[:, 0], X[:, 1], c=[colors[pred] for pred in k_means.labels_])
plt.scatter(k_means.cluster_centers_[:, 0], 
            k_means.cluster_centers_[:, 1], 
            c=colors, marker="^", s=100, edgecolor='k')
```
